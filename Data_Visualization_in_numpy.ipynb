```python
import numpy as np
import matplotlib.pyplot as plt
```

## 1. Line Plot




##### 1.1 Basic Line Plot

```python
x = np.linspace(1,10,100)
y=np.cos(x)
# print(x)
# print(y)
plt.plot(x,y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Basic Line Plot")
plt.show()
```

##### 1.2 Custom Line Graph

```python
x = np.linspace(1,10,50)
y=np.sin(x)
plt.plot(x,y,'p', linestyle='--', color = 'red', markersize= 5, alpha=.5)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Basic Line Plot")
plt.show()
```

##### 1.3 Line Plot With Multiple Lines


```python
x = np.linspace(1,10,50)
y1=np.sin(x)
y2=np.cos(x)
plt.plot(x,y1,'o', linestyle='--', color = 'red', markersize= 5, alpha=.5, label = 'sin(x)')
plt.plot(x,y2,'o', linestyle='--', color = 'blue', markersize= 5, alpha=.5,label = 'cos(x)')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Basic Line Plot")
plt.legend()
plt.show()
```

## 2. Scatter Plot

###### 2.1 Basic Scatter Plot

```python
x= np.random.rand(100)
y= np.random.rand(100)
plt.scatter(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Basic Scatter Plot')
plt.show()
```

##### 2.2 Customized Scatter Plot

```python
x= np.random.rand(500)
y= np.random.rand(500)
z= np.random.rand(500)
plt.scatter(x,y, marker='o',c=z,alpha=.7,cmap='viridis')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Basic Scatter Plot')
plt.colorbar()
plt.show()
```

##### 2.3 Scatter Plot with multiple classes

```python
x1= np.random.rand(200)-0.5
y1= np.random.rand(200)-0.5
x2= np.random.rand(200)+0.5
y2= np.random.rand(200)+0.5

plt.scatter(x1,y1, marker='o',alpha=.7)
plt.scatter(x2,y2, marker='p',alpha=.7)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Basic Scatter Plot')
# plt.colorbar()
plt.show()
```

##### 2.4 Hexbining Scatter Plot

```python
x= np.random.rand(1500)
y= np.random.rand(1500)
plt.hexbin(x,y,gridsize=20)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Hexbining Plot')
plt.colorbar()
plt.show()
```

## 3. Bar Graph

##### 3.1 Basic Bar Graph

```python
x=np.array(['A','B','C','D'])
y=np.array([10,20,30,40])
plt.bar(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Basic Bar Graph')
plt.show()
```

##### 3.2 Stacked Bar Graph

```python
data = np.random.rand(10,5)
# plt.bar(np.arange[10],data[:0],)
plt.bar(np.arange(10), data[:,0] + data[:,1] + data[:,2] + data[:,3] + data[:,4], label = 'D-5')
plt.bar(np.arange(10), data[:,0] + data[:,1] + data[:,2] + data[:,3], label = 'D-4')
plt.bar(np.arange(10), data[:,0] + data[:,1] + data[:,2], label = 'D-3')
plt.bar(np.arange(10), data[:,0] + data[:,1], label = 'D-2')
plt.bar(np.arange(10), data[:,0], label = 'D-1')

plt.legend()
plt.show()
```

##### 3.3 Grouped Bar Graph

```python
data = np.array([[10,20,30],
                 [40,50,60],
                 [70,80,90]])

width = .25
x = np.arange(3)

plt.bar(x - width, data[:,0], width,label = 'D-1')
plt.bar(x        , data[:,1], width, label = 'D-2')
plt.bar(x + width, data[:,2], width, label = 'D-3')

plt.legend()
plt.show()
```

### 4. Histogram

##### 4.1 Basic Histogram

```python
data=np.random.randn(1000000)
plt.hist(data,bins=1000)
plt.show()
```

##### 4.2 Cumulative Histogram

```python
data=np.random.randn(1000)
plt.hist(data,bins=10)
plt.show()
data=np.random.randn(1000)
plt.hist(data,bins=10,cumulative=True)
plt.show()
```

##### 4.3 Histogram with multiple Datasets

```python
data1=np.random.randn(10000)
data2=np.random.randn(10000) + 5
plt.hist(data1,bins=50,alpha=0.7,label='D1')
plt.hist(data2,bins=50,alpha=0.7,label='D2')
plt.legend()
plt.show()
```

### 5. Box Plot

```python
data = np.random.randn(100,5)
plt.boxplot(data, labels = ['D-1','D-2','D-3','D-4','D-5'], vert = False)
plt.show()
```

### 6. Heatmap

```python
data = np.random.rand(10,10)
plt.imshow(data)
plt.colorbar()
plt.show()
```

### 7. Pie Chart

```python
labels = ['Python', 'C++', 'Java', 'JavaScript', 'PHP']
sizes =  [21.3,        14.6 , 9.2,     8.5,       7.7]
explode = (0.12,        .1,    0,       0,           0)
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightpink']


plt.title("Programming Language Popularity")
plt.pie(sizes, colors=colors, labels=labels, autopct='%1.0f%%', shadow=True, explode=explode ,startangle=140)

plt.axis('equal')
plt.legend(title="Languages", loc="upper left")

plt.show()
```
